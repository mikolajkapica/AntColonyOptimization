```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("ants-data1.csv")
plt.title("ACO: 600/1/1/0.95/5")
plt.plot(df["iteration"], df["avg_dist"])
plt.plot(df["iteration"], df["min_dist"])
plt.plot(df["iteration"], df["max_dist"])
plt.legend(["avg_dist", "min_dist", "max_dist"])
plt.xlabel("iteration")
plt.ylabel("distance")
plt.savefig("ants-data-1.png", dpi=300)
plt.show()
```
